```python
import numpy as np
import pandas as pd
import os
```

```python
data_files = [i for i in os.listdir('updated_w_overall_counts') if 'il34' in i]
```

```python
df = pd.read_excel(os.path.join('updated_w_overall_counts', data_files[0]), index_col = 0)
```

```python
df.loc['Total Number of Spots'][0]
```

```python
filename = []
total_il34 = []

for file in data_files:
    print(file)
    df = pd.read_excel(os.path.join('updated_w_overall_counts', file), index_col = 0)
    
    total_il34 = np.append(total_il34, df.loc['Total Number of Spots'][0])
    filename = np.append(filename, file.split(' ')[0])
```

```python
pd.DataFrame(total_il34, index = filename).to_csv('total_il34_output.csv')
```
